```python
import random
import numpy as np

def match(k,num_matchs, days):
  acum = [0]*days
  for _ in range(k):
    day = random.choice(range(days))
    acum[day] += 1
  return max(acum) >= num_matchs
```

```python
def game(k,days,trials):
  contador = 0
  for _ in range(trials):
    contador += int(match(k,2,days))
  return contador/trials
```

```python
game(25,360,100000)
```

Usando Factoriales y Combinatoria


```python
import math

def permutation(days,k):
  return math.factorial(days)/math.factorial(days-k)

def paradox(days,k):
  return 1-(permutation(days,k)/(days**k))
```

```python
paradox(360,25)
```
