```python
import wbgapi as wb
import pandas as pd
```

```python
economies = wb.economy.DataFrame().reset_index() # This is very important.
economies
```

```python
#'NY.GDP.PCAP.PP.KD':'PBI per capita', 'SP.POP.GROW':'%PIB','NY.GDP.PCAP.PP.KD':'tasa de crecimiento poblacional'
df =wb.data.DataFrame(["NY.GNS.ICTR.ZS", "SP.POP.GROW", "NY.GDP.PCAP.PP.KD" ], time = range(2010, 2020), labels=True,
                  columns='series',
                  numericTimeKeys=True).reset_index()
df
```

```python
df.rename(columns={'NY.GDP.PCAP.PP.KD':'PBI per capita','NY.GNS.ICTR.ZS':'%PIB','SP.POP.GROW':'tasa de crecimiento poblacional'}, inplace= True)
df
```
