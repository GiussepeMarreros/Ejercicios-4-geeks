```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/precios-cantidades.csv')
df
```

```python
df.columns
```

```python
df.rename(columns={'Costos totales':'CostosTotales2010'},inplace=True)
df
```

```python
for year in range(2011,2014):
  df['tasa_crecimiento'+ str(year)]= (df['CostosTotales' + str(year)]-df['CostosTotales' + str(year-1)])/df['CostosTotales' + str(year-1)]

df
```

```python
df['tasa_media']= df[['tasa_crecimiento2011','tasa_crecimiento2012','tasa_crecimiento2013']].mean(axis=1)
df
```

```python
for year in range(2010,2014):
  df['ingresos'+str(year)]= df['precios'+str(year)]*df['cantidades'+str(year)]
  df['beneficio'+str(year)]= df['ingresos'+str(year)]-df['CostosTotales'+str(year)]

df
```

```python
precios=[]
for year in range(2010,2014):
  precios.append('precios'+str(year))
precios
```

```python
df['precios_promedio'] = df[precios].mean(axis=1)
df
```
