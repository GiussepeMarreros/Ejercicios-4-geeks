```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = []
y = []
acum  = 0
colors = []
rain = 10000
for _ in range(rain):
  x.append(random.uniform(-1,1))
  y.append(random.uniform(-1,1))
  if x[_]**2 + y[_]**2 <= 1:
    acum = acum + 1
    colors.append('c')
  else:
    colors.append('b')

print((acum/rain)*4)
```

```python
plt.scatter(x,y,color=colors)
```

**# Tarea:** Simular la probabilidad de 2 de los 8 punketos compartiendo fecha de cumpleaos (recordar formula y plantear como la probabilidad de que NO compartan para restrselo a 1 y obtener a as la probabilidad de que SI compartan)


```python
N = 365 #espacio muestral de fechas posibles
n = 25 #cantidad de punketos
acum = 1 #probabilidad de obtener un resultado unico en la primera sacada

for i in range(1, n):
    acum = acum*((N-i)/(365)) #la probabilidad de sacar un dia diferente a los ya obtenidos, va disminuyendo a medida que el valor de i aumenta

print(1-acum)
```
