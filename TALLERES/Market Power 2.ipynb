```python
#coheficiente de gini
def gini(array):
  def cumsum(array):
    output = [0]
    counter = 0
    total = sum(array)
    for e in array:
      counter = counter + e
      output.append(counter/total)
    return output
  array.sort()
  n = [i/len(array) for i in range(0,len(array)+1)]
  acum_array = cumsum(array)
  area = 0
  for i in range(1,len(acum_array)):
    area = area +  (acum_array[i] + acum_array[i-1]) * (n[i] - n[i-1])
  return 1  - area
```

```python
gini(df[df['sector']=='Software']['cm'].tolist())
```

```python
df['prod_sector']= df.groupby('sector')['q'].transform('sum')
df
```

```python
#% de participacion de cada empresa respecto de su sector
df['%participacion']=(df['q']/df['prod_sector'])*100
df
```

```python
def cv(serie):
    return serie.std()/serie.mean()
```

```python
def ratio(serie):
    return serie.max()/serie.min()
```

```python
#Calculo de medidas de tendencia central y dispercion de la produccion (q) y los costos (cm)
df.groupby(['sector'])[['q','cm']].agg(['mean','std','min',cv,ratio])
```

```python
#3 empresas con mayor y menor produccion  por cada sector
for sector in df['sector'].unique():
    print(sector)

    display(df[df['sector']== sector][['%participacion']].sort_values(by=['%participacion'],ascending=False,)[0:3])
    display(df[df['sector']== sector][['%participacion']].sort_values(by=['%participacion'],ascending=True,)[0:3])
```

```python
#empresas con los menores costos marginales(CM)
for sector in df['sector'].unique():
    print(sector)

    display(df[df['sector']==sector][['cm']].sort_values(by=['cm'],ascending=True)[0:3])
```
