```python
import numpy as np
import matplotlib.pyplot as ptl
import random
```

```python
def monty_hall_sim(change_choice, simulations=100000):
  wins = 0
  for _ in range(simulations):
    doors= [0, 0, 0] #3 puertas
    winner_door = random.randint(0,2) #se escoge una puerta ganadora
    doors[winner_door] = 1 #a la puerta escogida se le da el valor 1
    #el concursante elige una puerta al azar
    player_choice = random.randint(0,2) #el concursante escoje una puerta

    #el anfitrion abre una puerta que no tiene el premio
    possible_doors_to_open = [i for i in range(3) if i!=player_choice and doors[i] == 0]
    door_opened_by_host = random.choice(possible_doors_to_open)
    if change_choice:
      player_choice = [i for i in range(3) if i != player_choice and i != door_opened_by_host][0]

    if doors[player_choice] == 1:
      wins += 1

  win_percentage = (wins /simulations)*100
  return win_percentage

probabilidad_de_ganar_sin_cambio = monty_hall_sim(change_choice=True)
print(f"Probabilidad de ganar al cambiar de puerta: {probabilidad_de_ganar_sin_cambio:.2f}%")
```
