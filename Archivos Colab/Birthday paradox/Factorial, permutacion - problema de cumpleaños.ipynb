```python
def factorial(n):
  acum  = 1
  for value in range(2,n+1):
    acum = acum * value
  return acum
```

```python
factorial(4)
```

```python
factorial(5)
```

```python
def permutation(n,k):
  return factorial(n)/factorial(n-k)
```

```python
permutation(4,4)
```

```python
def combination(n,k):
  return permutation(n,k)/factorial(k)
```

```python
combination(4,4)
```

```python
def paradox(days, punketos):
  no_prob = permutation(days, punketos)/(days**punketos)
  return 1 - no_prob
```

```python
paradox(365, 25)
```
