```python
df  = pd.DataFrame({'var1':['rouge', 'rossa', 'red', 'blue', 'blua', 'yellow', 'jaune']})
```

```python
colores = {'rojo':['rouge', 'rossa', 'red'],
           'azul': ['blue', 'blua'],
           'amarillo':['yellow', 'jaune',]}
```

```python
df['var1_sph'] = ''
for key in colores.keys():
  for indice in df.index:
    if df.loc[indice, 'var1'] in colores[key]:
      df.loc[indice, 'var1_sph'] = key
```

```python
for color in df['var1_sph'].unique():
  df[color] = 0
  for index in df.index:
    reg_color = df.loc[index, 'var1_sph']
    df.loc[index, reg_color] = 1
```

```python
df = pd.DataFrame({'type':['micro digital',
                           'DIGITAL',
                           'empleado digital',
                           'empleado', 'productivo', 'libreinversion', 'libreinversion digital']})
```

```python
map_regex = {
'Digital' : r'(?<!libreinversion )digital',
'microcredito' : r'micro'
}
```

```python
#regNewVar
target = 'type'
col_name = 'Digital'
df[col_name] = 0
for indice in df.index:
  if bool(re.search(map_regex[col_name], df.loc[indice, target], re.IGNORECASE)):
    df.loc[indice, col_name] = 1
```
