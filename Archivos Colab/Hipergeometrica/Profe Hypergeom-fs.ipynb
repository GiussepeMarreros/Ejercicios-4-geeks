```python
def factorial(n):
  if n == 0:
    return 1
  else:
    acum = n
    for num in range(1,n):
      acum = acum*num
  return acum
```

```python
def comb(n,k):
  num = factorial(n)
  den = factorial(n-k) * factorial(k)
  return num/den
```

```python
def hypergeom(k, n, M, N):
  num =  comb(n,k) * comb(M -n, N-k)
  den = comb(M,N)
  return num/den
```

```python
k = 8
n = 13
M = 25
N = 12
```

```python
hypergeom(k, n, M,N)
```
