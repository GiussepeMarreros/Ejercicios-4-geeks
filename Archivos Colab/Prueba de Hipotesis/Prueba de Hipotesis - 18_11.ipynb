```python
mu=7
std = 1.5
pop= normal(mu,std, 1230)
```

```python
means= []
trials = 1000
size = 45
for i in range(trials):
  means.append(np.random.choice(pop,size).mean())
```

```python
plt.hist(means)
```

```python
si=[]
for media in means:
  if media>7.5:
    si.append(1)
```

```python
len(si)/trials
```
