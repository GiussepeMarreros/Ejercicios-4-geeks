```python
# uniformizar valores
convert_dict= {'var1':float,'var2':bool, 'var3':'category','var4':float, 'var5':float,'var6':'category'}
categories_dict={'var3':['rojo','verde','azul','blanco'],'var6':['acepta','no acepta']}

def uniformizar(df):
    df=df.replace(['Nan','NAN','No registra','@'],np.nan)
    for col in df.columns:
        if convert_dict[col]==float or convert_dict[col]==bool:
            df[col]=df[col].replace('o','0',regex=True)
            df[col]=df[col].replace(' ','',regex=True)
        else:
            df[col]=df[col].str.lower()
            df[col]=df[col].replace('\d+','',regex=True)
            for value in df[col].values:
                if value not in categories_dict[col]:
                    df[col]=df[col].replace(value,np.nan)
    df=df.astype(convert_dict)
    return df

df = uniformizar(df)
df
```

```python
zvar1= []
for i in df['var1']:
    calculo=(i-(df['var1'].mean()))/df['var1'].std()
    zvar1.append(calculo)
zvar1
```

```python
df['zvar1']=zvar1
df
```

```python
zvar4= []
for j in df['var4']:
    calculo2=(j-(df['var4'].mean()))/df['var4'].std()
    zvar4.append(calculo2)
zvar4
```

```python
df['zvar4']=zvar4
df
```
