```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/icfesSaber11_20202.csv"
df = pd.read_csv(url)
```

```python
df = df[df['COLE_DEPTO_UBICACION']=='BOGOT√Å'].copy()
```

```python
mean = df['PUNT_MATEMATICAS'].mean()
std = df['PUNT_MATEMATICAS'].std() #sigma
norm.interval(0.95,loc=mean,scale= std/((df['PUNT_MATEMATICAS'].shape[0])**0.5))
```
