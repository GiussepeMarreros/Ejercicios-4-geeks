```python
pob = np.random.normal(22,15.2,300)
pob
```

```python
np.array([10,13,5]).std()
```

```python
trials = 10000
sesgo = []
nosesgo = []
sample_size = 60
for _ in range(trials):
  sample = np.random.choice(pob,sample_size,replace=False)
  sesgo.append(sample.std())
  nosesgo.append(sample.std(ddof=1))
```

```python
np.array(sesgo).mean(),np.array(nosesgo).mean()
```

```python
pop_size=1000
mu, sigma = 80.0, 4.78
```

```python
pop = np.random.normal(loc = mu, scale=sigma, size=pop_size)
pop
```

```python
trials = 99999
sample_size= 15
biased, unbiased=[], []
for _ in range(trials):
  sample = np.random.choice(pop, sample_size, replace=False)
  biased.append(sample.std(ddof=0))
  unbiased.append(sample.std(ddof=1))
```

```python
np.array(biased).mean(), np.array(unbiased).mean() #el unbiased esta mas cerca al estimado verdadero
```
