```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from statsmodels.stats.proportion import proportion_confint
```

```python
url = 'https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/basesuper.csv'
df = pd.read_csv(url)
```

```python
mean = df['Ganancia (pérdida)'].mean()
std = df['Ganancia (pérdida)'].std()

print(f"{mean:.2f}")
print(f"{std:.2f}")


```

```python
vars = ['Activos corrientes totales','Pasivos corrientes totales']
for var in vars:
  for sector in df['sector'].unique():
    mean = df[df['sector'] == sector][var].mean()
    sigma = df[df['sector'] == sector][var].std()
    n=df[df['sector'] == sector].shape[0]
    print(var, sector)
    print(mean, norm.interval(loc=mean, scale=sigma/(n**0.5), confidence=0.95))
```

Razon Corriente

```python
df['razon'] = df['Activos corrientes totales']/df['Pasivos corrientes totales']
razon= df.groupby(['sector'])['razon'].mean()
print(razon)
```

```python
# activos corriente o circulantes entre pasivos corrientes
#Activos corrientes totales
#Pasivos corrientes totales
mean_ac = df['Activos corrientes totales'].mean()
mean_pc = df['Pasivos corrientes totales'].mean()
razon_corriente = mean_ac/mean_pc
print(razon_corriente)
#quiere decir que con el activo corriente, que tiene mas liquidez, se puede pagar 1.66 veces la deuda corriente de la empresa
```

ROA

```python
df['roa'] = df['Ganancia (pérdida)']/df['Total de activos']
roa= df.groupby(['sector'])['roa'].mean()
print(roa)
```

```python
#Retorno sobre los activos de todas las empresas
mean_return = df['Ganancia (pérdida)'].mean()
mean_actives = df['Total de activos'].mean()
roa = mean_return/mean_actives
print(roa)
```

ROE

```python
df['roe'] = df['Ganancia (pérdida)']/df['Patrimonio total']
roe= df.groupby(['sector'])['roe'].mean()
print(roe)
```

```python
#Retorno sobre los activos de todas las empresas
mean_return = df['Ganancia (pérdida)'].mean()
mean_equity = df['Patrimonio total'].mean()
roe = mean_return/mean_equity
print(roe)
```
