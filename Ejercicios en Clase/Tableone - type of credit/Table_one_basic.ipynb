```python
import numpy as np
import pandas as pd
from tableone import TableOne
```

```python
sesgo = 20
baseline = 10
size  = 100
edada = pd.Series(np.random.normal(baseline,2.5, size))
edadb = pd.Series(np.random.normal(baseline + sesgo, 1, size))
pd.DataFrame()
df  =  pd.concat([edada, edadb])
df = pd.DataFrame(df, columns=['edad'])
df['group'] = ['A' if x < size else 'B' for x in range(size*2)]
df.reset_index(inplace=True, drop=True)
TableOne(df,  groupby='group', pval=True, htest_name=True)
# P < 0.05 existen diferencias significativas entre los grupos y la varaible de intereés
```

```python
edadc  =  pd.DataFrame(pd.Series(np.random.normal(baseline + sesgo, 1, size)), columns=['edad'])
df= pd.concat([df, edadc])
df['group'] = np.where(pd.isnull(df['group']), 'C', df['group'])
df.reset_index(inplace=True, drop=True)
TableOne(df,  groupby='group', pval=True, htest_name=True)
```

```python
sex = np.random.choice(["F","M"], p=[0.1, 0.9], size=size*3)
df['sex']  = sex
```

```python
df
```

```python
TableOne(df, categorical=['sex'], groupby='group', pval=True, htest_name=True)
```

```python
# cuando p > 0.05 no están relacionados es decir son independientes.!
# cuando p < 0.05 están relacionados o son dependientes!!!
```

```python
df['days'] = np.random.exponential(5, 300)
df['group2'] = np.where((df['group']=='A') | (df['group']=='B'), 'cond1', 'cond2')
```

```python
TableOne(df, categorical=['sex', 'group2'], nonnormal=['days'], groupby='group', pval=True, htest_name=True)
```

```python
TableOne(df, categorical=['sex', 'group'], nonnormal=['days'], groupby='group2', pval=True, htest_name=True)
```
