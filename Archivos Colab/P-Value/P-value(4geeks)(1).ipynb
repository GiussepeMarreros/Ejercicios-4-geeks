```python
mu = 305
sd = 120
# según estudio que hicieran en chicago la media de la población de ingresos es de 305
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
from scipy.stats import norm
```

# testear el estudio la hipotesis
 $$ H_{0} : \mu = 305 $$
 $$ H_{1} : \mu \neq 305 $$

```python
# hice sample de 78 personas y obtuve un promedio 365
```

```python
# si realmente mu es 305 cual es la probabilida de observar un promedio
# en una muestra de 78 personas de 321
```

```python
trials = 100000
size = 78
means = []
for _ in range(trials):
  sample = normal(mu,sd, size)
  means.append(sample.mean())
```

```python
plt.hist(means)
```

```python
sum([1 for mean in means if mean > 321])/trials
```

```python
1 - norm.cdf(321, loc = mu, scale = sd/size**0.5)
```
