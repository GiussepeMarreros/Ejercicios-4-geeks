```python
def std_str(string):
  if isinstance(string, str):
    string = string.upper()
    string = string.replace("Á", "A")
    string = string.replace('É', 'E')
    string = string.replace('Í', 'I')
    string = string.replace('Ó', 'O')
    string = string.replace('Ú', 'U')
    return string
  else:
    return string
```

```python
df['type_of_credit'] = df['type_of_credit'].apply(std_str)
df
```
