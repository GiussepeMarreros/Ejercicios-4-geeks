$$ f(x) = 120x - 4x^{2} $$

$$ 120 - 8 x $$
$$ x  = 120/8$$
$$ x = 15.0$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def quad(x):
  return 120*x - 4*x**2
```

```python
def gradient(x):
  return -1*(120 - (8*x))
```

```python
optima = 15
```

```python
domain = np.linspace(5,25, 200)
y = [-quad(x) for x in domain]
plt.plot(domain, y)
plt.axvline(optima, color='red')
```

$$ w_{t+1} = w_{t} - \alpha \frac{\partial J (w_{t})}{\partial w}$$

```python
updates = 10
wcurrent = 44
alpha  = 0.1
for _ in range(updates):
  wnext = wcurrent - alpha * gradient(wcurrent)
  wcurrent = wnext
  print(wcurrent)
```
