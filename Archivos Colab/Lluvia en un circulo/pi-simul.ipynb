```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
X = []
Y = []
rain = 10000000
for _ in range(rain):
  X.append(random.uniform(-1,1))
  Y.append(random.uniform(-1,1))
acum  = 0
for x,y in zip(X,Y):
  if x**2 + y**2 <= 1:
    acum = acum + 1
print((acum/rain)*4)
```
