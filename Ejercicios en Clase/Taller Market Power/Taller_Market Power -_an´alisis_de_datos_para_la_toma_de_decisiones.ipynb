```python
df['prod_sector']=df.groupby('sector')['q'].transform('sum')
df['%participacion'] = (df['q']/df['prod_sector'])*100
df
```

```python
def gini(array):
  def cumsum(array):
    output = [0]
    counter = 0
    total = sum(array)
    for e in array:
      counter = counter + e
      output.append(counter/total)
    return output
  array.sort()
  n = [i/len(array) for i in range(0,len(array)+1)]
  acum_array = cumsum(array)
  area = 0
  for i in range(1,len(acum_array)):
    area = area +  (acum_array[i] + acum_array[i-1]) * (n[i] - n[i-1])
  return 1  - area
```

```python
#coheficiente de gini calcula en este caso la igualdad en la distribucion de los "costos Marginales" entre las empresas del "sector Software"
gini(df[df['sector']=='Software']['cm'].tolist())
```

```python
def cv(serie):
    return serie.std()/serie.mean()
```

```python
def ratio(serie):
    return serie.max()/serie.min()
```

```python
#Medida de tendencia central para la produccion(q) y el costo (cm)
df.groupby(by=['sector'])[['q','cm']].agg(['mean','std','min',cv,ratio])
#print("Mediana de la produccion (q) por sector:",df_mediana_produccion)
```
