```python
from scipy.stats import norm
from numpy.random import normal
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 45
std = 23
pop = normal(mu, std, 1230)
```

```python
means = []
trials = 50000
size = 45
for _ in range(trials):
  means.append(np.random.choice(pop, size ).mean())
```

```python
plt.hist(means)
```

```python
si = []
for media in means:
  if media>51:
    si.append(1)
```

```python
len(si) / trials
```
