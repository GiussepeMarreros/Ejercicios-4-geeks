```python
#frecuencia de las columnas (cuantas veces aparece un valor en una columna)
#moda es escoger el valor que mas se repite

#algunas funciones no requieren parametros oblicatoriosy funcionan solo con (), pero otras si requieren parametros para funcionar
```

```python
#parametro -- by=var1 es la variable, contiene dominio porque hay valores que se repiten, en var 3 o 4 no hay valores dominio. agrupar entre
#parametro -- level es que resultado quiero tener de la agrupacion solicitada con by=var1 (QUE ES LO QUE QUIERES CONTAR)
#df.groupby(by='var1') solo es una identificacion, count() es el comando para ejecutar una accion 
df_filtrado=df.groupby(by='var1').count()
df_filtrado
```

```python
#groupby me sirve cuando la columna agrupada tiene pocos valores unicos (ej:male, female(2))
df_filtrado1=df.groupby(by='var2').value_counts()
df_filtrado1
```

```python
#size da el tamaño del grupo en este caso

df_filtrado2=df.groupby(by='var1').size()
df_filtrado2
```

```python
#Los parametros estan en parentesis, pero si quiero agrupar en una variable adicional pongo entre llaves
df_filtrado3=df.groupby(by=['var1','var2']).value_counts()
df_filtrado3
```

```python
#se puede seguir agrupando en orden mas variables adicionales
df_filtrado4=df.groupby(by=['var1','var2','var4']).mean()
df_filtrado4
```

```python
#solo toma var4 para hacer el promedio, agrupando por var1 y var2
#casi siempre se usa para sacar una estadistica, medidas de tendencia central(promedio, conteo, std etc.)
df_filtrado4=df.groupby(by=['var1','var2'])['var4'].mean()
df_filtrado4
```

```python
df_filtrado4=df.groupby(by=['var1','var2'])['var4'].std()
df_filtrado4
```
