```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from statsmodels.stats.proportion import proportion_confint
```

```python
url  = "https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/basesuper.csv"
df = pd.read_csv(url)
df
```

```python
df['sector']  = df['Clasificación Industrial Internacional Uniforme Versión 4 A.C'].apply(lambda row: row[0:1])
```

```python
variables = ['var1', 'var2']
sectores = ['a', 'b', 'c']
for var in variables:
  for sector in sectores:
    print(var, sector)
```

```python
for sector in df['sector'].unique():
  print(sector)
```

```python
df['sector'].value_counts()
```

```python
vars = ['Ganancia (pérdida)',  'Total de patrimonio y pasivos']
for var in vars:
  for sector in df['sector'].unique():
    mean = df[df['sector']==sector][var].mean()
    sigma = df[df['sector']==sector][var].std()
    n = df[df['sector']==sector].shape[0]
    print(var, sector)
    print(mean, norm.interval(loc=mean, scale = sigma/(n**0.5), confidence=0.95))
```

```python

# Confidence interval mean
print(norm.interval(loc = 12, scale=4, confidence=0.95))
# Confidence interval proportion
proportion_confint(30, 130, 0.05)
```
