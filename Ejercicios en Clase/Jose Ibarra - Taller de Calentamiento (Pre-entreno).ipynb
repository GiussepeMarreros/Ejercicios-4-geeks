```python
def df_resumen_normal(metrica):
  dict_0 = {'sector': [],
        'media': [],
        'std': [],
        'ci_lim_inf': [],
        'ci_lim_sup' : []}

  for sector in df['grupos'].unique():
    media = df[df['grupos']==sector][metrica].mean()
    std = df[df['grupos']==sector][metrica].std()
    dict_0['sector'].append(sector)
    dict_0['media'].append(media)
    dict_0['std'].append(std)
    dict_0['ci_lim_inf'].append(stats.norm.interval(alpha, loc = media, scale = std)[0])
    dict_0['ci_lim_sup'].append(stats.norm.interval(alpha, loc = media, scale = std)[1])
  final_df = pd.DataFrame.from_dict(dict_0)
  return final_df

df_resumen_normal("Ganancia (pérdida)")
```

```python
def df_resumen_proporcion_binom(evento):
  dict_0 = {'sector': [],
        'p': [],
        'total': [],
        'quiebras': [],
        'error_estandar': [],
        'ci_lim_inf': [],
        'ci_lim_sup' : []}

  for sector in df['grupos'].unique():
    p = df[df['grupos']==sector][evento].mean()
    n = df[df['grupos']==sector][evento].count()
    n_exitos = df[(df['grupos']==sector) & (df['y']== 1)][evento].count()
    ci = sm.proportion.proportion_confint(n_exitos, n, alpha)
    error = p*(1-p)/np.sqrt(n)

    dict_0['sector'].append(sector)
    dict_0['p'].append(str(round((p*100),2))+'%')
    dict_0['total'].append(n)
    dict_0['quiebras'].append(n_exitos)
    dict_0['error_estandar'].append(error)
    dict_0['ci_lim_inf'].append(str(round((ci[0]*100),2))+'%')
    dict_0['ci_lim_sup'].append(str(round((ci[1]*100),2))+'%')
  final_df = pd.DataFrame.from_dict(dict_0)
  return final_df

df_resumen_proporcion_binom('y')
```

```python
def ci(sample, z):
  n = len(sample)
  p_hat = sum(sample)/n
  se = np.sqrt(p_hat*(1-p_hat)/n)
  ci_lim_inf = p_hat - z*se
  ci_lim_sup = p_hat + z*se
  return ci_lim_inf, ci_lim_sup
```
