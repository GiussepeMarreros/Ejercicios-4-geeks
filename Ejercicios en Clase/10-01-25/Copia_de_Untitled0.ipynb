```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/it-ces/Datasets/refs/heads/main/basesuper.csv')
```

```python
df.info()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df.drop(col, inplace=True, axis=1)

df.info()
```

```python
for col in df. columns:
  if df[col].isnull().sum() > 12773*0.3:
    df.drop(col, axis=1, inplace=True)

df.info()
```

```python
df['razon_corriente'] = df['Activos corrientes totales']/df['Pasivos corrientes totales']
```

```python
df['ROE'] = df["Ganancia (pérdida)"]/df['Patrimonio total']
```

```python
df['Ingreso total'] = df['Ingresos de actividades ordinarias'] + df['Otros ingresos'] + df['Ingreso (gasto) por impuestos']
```

```python
df['ROA'] = df['Ingreso total']/df['Total de activos']
```

```python
df['ROE'] = df["Ganancia (pérdida)"]/df['Patrimonio total']
```

```python
'''
# Tomamos conceptos relacionados con ingresos , , gastos y elementos financienss clave para las ulitidades
‘’’ Variables que impactan a la utilidad
1.	Ingresos de actividades ordinarias: Representan los ingresos principales de la empresa por sus operaciones básicas. Un aumento en estos ingresos suele incrementar las utilidades.
2.	Costo de ventas: Es el costo directo asociado con la producción de los bienes vendidos por la empresa. Un aumento en el costo de ventas sin un aumento correspondiente en los ingresos puede reducir las utilidades.
3.	Ganancia bruta: Es la diferencia entre los ingresos de actividades ordinarias y el costo de ventas. Este es un indicador importante de la eficiencia operativa y el impacto directo en las utilidades.
4.	Otros ingresos: Representan ingresos adicionales no relacionados con las operaciones principales de la empresa, que pueden aumentar las utilidades.
5.	Gastos de ventas: Gastos relacionados con la distribución y venta de productos. Un aumento en estos gastos sin un aumento correspondiente en los ingresos puede reducir las utilidades.
6.	Gastos de administración: Son los costos relacionados con la administración general de la empresa. Un manejo eficiente de estos gastos es crucial para mantener altas las utilidades.
7.	Otros gastos: Gastos adicionales que no están directamente relacionados con las operaciones principales de la empresa. Reducción en estos gastos puede mejorar las utilidades.
8.	Otras ganancias (pérdidas): Incluyen ganancias o pérdidas no relacionadas con las operaciones principales, que pueden afectar las utilidades netas.
9.	Ganancia (pérdida) por actividades de operación: Es una medida clave que se obtiene restando todos los gastos operativos de la ganancia bruta. Representa la rentabilidad de las operaciones principales antes de considerar ingresos y gastos financieros.
10.	Ingresos financieros: Ingresos por inversiones financieras como intereses y dividendos. Un aumento en estos ingresos puede incrementar las utilidades.
11.	Costos financieros: Intereses y otros gastos relacionados con financiamiento. Un aumento en estos costos puede reducir las utilidades netas.
12.	Ganancia (pérdida) antes de impuestos: Representa las utilidades de la empresa antes de la deducción de impuestos. Es un indicador crucial del margen de beneficio.
13.	Ingreso (gasto) por impuestos: Los impuestos que la empresa debe pagar sobre sus ganancias, los cuales afectan directamente las utilidades netas.
14.	Ganancia (pérdida) procedente de operaciones continuadas: Representa las utilidades de la empresa antes de considerar operaciones no continuadas, lo cual es crucial para medir el desempeño sostenible de la empresa.
15.	Ganancia (pérdida) procedente de operaciones discontinuadas: Representa ingresos o pérdidas de operaciones que ya no forman parte de las actividades principales de la empresa.
16.	Ganancia (pérdida): La utilidad neta final, que resume el impacto de todos los factores, tanto operativos como financieros, sobre la rentabilidad final de la empresa  ‘’’
'''
```

```python
correlation_matrix = df[["Ingresos de actividades ordinarias", "Costo de ventas", "Ganancia bruta", "Otros ingresos","Gastos de administración", "Otros gastos", "Ganancia (pérdida) por actividades de operación","Costos financieros", "Ganancia (pérdida) antes de impuestos", "Ingreso (gasto) por impuestos", "Ganancia (pérdida) procedente de operaciones continuadas",  "Ganancia (pérdida)"]].corr()
filtered_corr = correlation_matrix.copy()
filtered_corr[np.abs(filtered_corr) <= 0.5] = np.nan
fig, axis = plt.subplots(figsize=(10, 7))
sns.heatmap(filtered_corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, linewidths=.5, linecolor='black')
plt.tight_layout()
plt.show()
```
